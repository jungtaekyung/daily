# group by
  + 데이터를 특정 컬럼으로 묶어서 그룹화
  + 아래의 세 단계를 적용하여 데이터를 그룹화(groupping) (SQL의 group by 와 개념적으로는 동일, 사용법은 유사)
    - 데이터 분할 (split)
    - operation 적용 (apply)
    - 데이터 병합 (combine)

```python
#↑ 이를 split-apply-combine strategy 라고 함
# 데이터를 활용한 프로세싱의 기본 접근 방법
```

```python
import math
def age_categorize(age):
    if math.isnan(age): 
        return - 1    #Age가 없으면 -1로처리
    return int(age/10)*10
```

```python
Age_span = df['Age'].apply(age_categorize) 
df['Age_Span'] = Age_span
```

```python
grouped_pclass = df.groupby('Pclass')
grouped_pclass #DataFrameGroupBy객체 
#groupby 객체의 속성
# - groups -> dict 형태
# - key -> 컬럼의 각 값
# - value -> 인덱스 값들의 list
```

## index를 이용한 group by
 - index가 있는 경우, groupby 함수에 level 사용 가능
   - level은 index의 depth를 의미하며, 가장 왼쪽부터 0부터 증가
   
* **set_index** 함수
 - column 데이터를 index 레벨로 변경
* **reset_index** 함수
 - 인덱스 초기화


```python
df.set_index('Pclass') #Pclass 를 index로 
```

```python
df.set_index(['Pclass','Sex']) #Pclass 와 Sex의 multi index
```

```python
df.set_index(['Pclass','Sex']).reset_index() #index 초기화 
#index에서 벗어난 Pclass 와 Sex는 데이터프레임의 가장 앞으로
```

```python
df.set_index('Sex').groupby(level=0).mean() #Sex를 인덱스화 하고 0번 level 인덱스별 group화
```

```python
df.set_index('Age').groupby(age_categorize).mean() # Sex를 인덱스화 하고 인덱스별 함수를 이용한 그룹화 
```

```python
df.set_index('Age_Span').groupby(level=0).mean() 
```

```python
#나이대별로 생존률
df.set_index('Age_Span').groupby(level=0).mean()['Survived']
#df.set_index('Age').groupby(age_categorize).mean()['Survived']
```

```python
df.set_index(['Pclass','Age_Span']).groupby(level=0).mean()  #Pclass, Age_Span multiindexing 후 level=0 인덱스별 그룹핑
```

```python
df.set_index(['Pclass','Age_Span']).groupby(level=1).mean()  #Pclass, Age_Span multiindexing 후 level=1 인덱스별 그룹핑
```

```python
df.set_index(['Pclass','Age_Span']).groupby(level=[0,1]).mean() #Pclass, Age_Span multiindexing 후 level=[0,1] 인덱스별 그룹핑
```

```python
df.set_index(['Pclass','Age_Span']).groupby(level=[1,0]).mean() #Pclass, Age_Span multiindexing 후 level=[1,0] 인덱스별 그룹핑
```

## aggregate(집계) 함수 사용하기
 - groupby 결과에 '집계함수'를 적용하여 그룹별 데이터 확인 가능


```python
df.groupby(['Pclass','Age_Span']).aggregate(np.sum)
```

```python
df.groupby(['Pclass','Age_Span']).aggregate([np.mean,np.sum,np.max])
```
