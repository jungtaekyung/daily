생존자 별 나이 평균으로 NaN값대체

```python
sur_mean0 = df[df['Survived'] == 0].Age.mean() #df[df['Survived'] == 0]['Age'].mean()
sur_mean0 
```

```python
sur_mean1 = df[df['Survived'] == 1].Age.mean() #df[df['Survived'] == 1]['Age'].mean()
sur_mean1 
```

```python
df.loc[df['Survived']==0,'Age'] = df[df['Survived']==0].Age.fillna(sur_mean0)
df.loc[df['Survived']==0,'Age']
#사망자 들 중 Age가 NaN인 데이터에 sur_mean0(사망자 나이평균) 채우기
```

```python
df.loc[df['Survived']==1,'Age'] = df[df['Survived']==1].Age.fillna(sur_mean1)
df.loc[df['Survived']==1,'Age']
#생존자 들 중 Age가 NaN인 데이터에 sur_mean1(생존자 나이평균) 채우기
```
