```python
from selenium import webdriver
```

```python
driver_path = r'C:\DevRoot\dataset\bin\chromedriver.exe'  # chromdriver 경로
driver = webdriver.Chrome(driver_path)                    # chromedriver와 selenium 연동
                                                          # chromedriver 실행 빈페이지
```

```python
driver.get('https://movie.naver.com/movie/sdb/rank/rmovie.naver')# chromedriver에 네이버 영화 실행 
```

```python
#css 로 element(들) 선택 -> Element 의 list 리턴
elements = driver.find_elements_by_css_selector('#old_content > table> tbody td.title a') #영화 랭킹 크롤링
len(elements) # 영화 랭킹 50개
```

```python
elements[0].text                              # 1위 영화 
```

```python
 elements[0].get_attribute('href')    # 특정 attribute 값  -> '절대주소' 로 리턴해줌
                                      # 1위 영화 디테일 페이지 링크
```

```python
elements[0].tag_name                   #태그 이름 -> a 태그
```

```python
elements[0].rect                      # element 의 좌표, 크기
```

```python
elements[0].value_of_css_property('display')   # css property 값!
```
