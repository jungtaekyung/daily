```python
import requests
from bs4 import BeautifulSoup
def crawl_daum_blog(search,page=1): #페이지 추가
    url = f'https://search.daum.net/search?w=blog&nil_search=btn&DA=NTB&enc=utf8&q={search}&p={page}'
    response = requests.get(url) #정상적인 리턴
    data = BeautifulSoup(response.text, "html.parser")
    post_elements = data.select("#blogColl .list_info li .wrap_cont")
    
    result=[
        {
            'post_title' : post_element.select_one('.wrap_tit').text.strip(),
            'url' : post_element.select_one('.f_link_b').attrs.get('href').strip(),
            'blog_title': post_element.select_one('.info .f_nb').text.strip(),
        }
        for post_element
        in post_elements
    ]
    return result
```

```python
crawl_daum_blog('adsp자격증',2)
```

```python
result = []
for page in range(1, 11):    # 1페이지부터 10페이지까지 크롤링
    print(f'{page} 페이지 크롤링...') #크롤링 할때마다 문장출력
    result += crawl_daum_blog("파이썬", page)  # 각페이지의 크롤링 한것을 쌓음
    
result
```
